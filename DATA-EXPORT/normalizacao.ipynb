```python
import pandas as pd

dados = pd.read_csv('cdd.csv', dtype=pd.StringDtype())
#dados.head()
```

```python
# quando: bla VER ble
# bla: remissiva
# ble: assunto

### vamos criar uma coluna nova chamada remissiva
```

```python
dados_remissiva = dados[dados['titulo'].str.contains(" VER ")]
#dados_remissiva['Remissiva'].unique()
```

```python
dados[['First','Last']] = dados.titulo.str.split(" VER ",expand=True,)
```

```python
for index, row in dados.iterrows():
    if isinstance(row['First'], str) and isinstance(row['Last'], str):
        dados.at[index,'titulo'] = row['Last']
        dados.at[index,'Remissiva'] = row['First']
dados
```
